```python
arr = [-6,6,6,-5,-10,3,-2,8,5,-5]

cur_sum = 0
max_sum = 0
cur_start = 0
max_start = 0

for i in range(len(arr)):
    cur_sum += arr[i]
    if cur_sum <= 0:
        cur_sum = 0
        cur_start = i+1
    elif cur_sum > max_sum:
        max_sum = cur_sum
        max_end = i
        max_start = cur_start

print(max_start, max_end, max_sum)
```
